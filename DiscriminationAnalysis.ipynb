# Chemical Discrimination Testing with Colorimetric Sensor Arrays

**Objective**:  
Investigate whether a colorimetric sensor array can reliably distinguish between different toxic industrial chemicals (TICs) based on 2-minute exposure measurements.

**Main Research Question**:  
*Are the inter-chemical differences significantly greater than the intra-chemical variability (between replicates)?*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy.stats import kruskal, mannwhitneyu, ks_2samp, spearmanr
from scipy.spatial.distance import pdist, squareform
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("DB_220_IDLH_2mins.txt", sep='\t', header=None)

df.rename(columns={0: 'sample'}, inplace=True)

data = df.set_index('sample').T

clean_data = data.apply(pd.to_numeric, errors='coerce').dropna(axis=1, how='all')

df = clean_data.T

df = df.reset_index()
df.head()
df.tail()
```

```python
df['Chemical'] = df['sample'].apply(lambda x: x.split('_')[0])
df['Replicate'] = df['sample'].apply(lambda x: '_'.join(x.split('_')[1:]))

sensor_columns = df.columns.difference(['sample', 'Chemical', 'Replicate'], sort=False)
df = df[['sample', 'Chemical', 'Replicate'] + list(sensor_columns)]

df.head()
```

```python
sensor_data = df.drop(columns=['sample', 'Chemical', 'Replicate'])

df_grouped = df[['Chemical']].join(sensor_data)

group_means = df_grouped.groupby('Chemical').mean()
overall_mean = df_grouped.drop(columns='Chemical').mean()

inter_group_var = np.mean(np.var(group_means, axis=0))
print(f"Inter-group variance (across chemical means): {inter_group_var:.4f}")

intra_group_vars = df_grouped.groupby('Chemical').apply(lambda g: np.mean(np.var(g.drop(columns='Chemical'), axis=0)))
intra_group_var = intra_group_vars.mean()
print(f"Average intra-group variance (within chemicals): {intra_group_var:.4f}")

print(f"Ratio (inter / intra): {inter_group_var / intra_group_var:.4f}")
```

The variance analysis reveals a clear and compelling pattern: the average inter-group variance across chemical means is approximately 483.20, while the average intra-group variance within replicates is just 6.18. This yields a striking inter/intra variance ratio of over 78, indicating that the sensor array’s responses to different chemicals are far more distinct than the natural variation among replicates. In practical terms, this strongly suggests that the sensor array has high discriminatory power, with distinct chemical identities producing reliably different sensor profiles.

Now that the variance analysis has demonstrated strong overall separation between chemical groups, the next phase of the analysis focuses on confirming the statistical significance of these differences and identifying where they occur. This begins with applying the Kruskal-Wallis H-test, a nonparametric alternative to ANOVA, to each sensor individually. The goal is to determine whether at least one chemical group differs significantly from the others in terms of sensor response. A low p-value (typically below 0.05) indicates that the sensor detects meaningful differences across chemical groups, supporting its role in chemical discrimination.

Following this, the analysis transitions to pairwise comparisons using the Mann-Whitney U-test. This test is applied between every pair of chemicals, across each sensor, to identify specifically which chemical pairs show statistically distinct responses. This step is essential for mapping out the array’s discriminatory capabilities and highlighting any chemical pairs that might be difficult to distinguish.

To further complement these findings, additional tests such as the Kolmogorov-Smirnov (KS) test may be used to assess differences in the distribution shapes of sensor responses — offering insight into how response variability differs across chemicals, beyond just central tendencies. Finally, Spearman correlation analysis can be used to examine the overall similarity of response patterns between chemicals, helping to detect clusters of chemically similar compounds.

In short, this phase deepens the analysis by moving from broad group-level differences to fine-grained pairwise and pattern-level comparisons, allowing for a comprehensive assessment of the sensor array’s performance.

```python
kruskal_results = {}

for sensor in sensor_data.columns:
    group_data = [group[sensor].values for name, group in df.groupby('Chemical')]

    stat, p = kruskal(*group_data)
    kruskal_results[sensor] = {'statistic': stat, 'p_value': p}

kruskal_df = pd.DataFrame.from_dict(kruskal_results, orient='index')
kruskal_df.sort_values('p_value', inplace=True)

import os

os.makedirs('results', exist_ok=True)

kruskal_df.to_csv('results/kruskal_result.csv', index=True)

print("Kruskal-Wallis results saved to: results/kruskal_result.csv")

kruskal_df.head()
```

```python
from itertools import combinations
from tqdm import tqdm

mw_results = []

for sensor in tqdm(sensor_data.columns, desc="Processing sensors"):

    grouped = df.groupby('Chemical')[sensor]
    
    for chem1, chem2 in combinations(grouped.groups.keys(), 2):
        data1 = grouped.get_group(chem1)
        data2 = grouped.get_group(chem2)
        
        stat, p = mannwhitneyu(data1, data2, alternative='two-sided')
        
        mw_results.append({
            'Sensor': sensor,
            'Chemical_1': chem1,
            'Chemical_2': chem2,
            'U_statistic': stat,
            'p_value': p
        })

mw_df = pd.DataFrame(mw_results)

os.makedirs('results', exist_ok=True)
mw_df.to_csv('results/mannwhitney_results.csv', index=False)

print("Mann-Whitney U-test results saved to: results/mannwhitney_results.csv")
mw_df.head()
```

```python
sensor_data_only = df.drop(columns=['sample', 'Chemical', 'Replicate'])

chemical_means = sensor_data_only.groupby(df['Chemical']).mean()

spearman_corr = chemical_means.T.corr(method='spearman')

os.makedirs('results', exist_ok=True)
spearman_corr.to_csv('results/spearman_correlation.csv')

print("Spearman correlation matrix saved to: results/spearman_correlation.csv")

spearman_corr.head()
```

```python
ks_results = []

for sensor in tqdm(sensor_data.columns, desc="Processing sensors for KS"):
    grouped = df.groupby('Chemical')[sensor]
    
    for chem1, chem2 in combinations(grouped.groups.keys(), 2):
        data1 = grouped.get_group(chem1)
        data2 = grouped.get_group(chem2)
        
        stat, p = ks_2samp(data1, data2)
        
        ks_results.append({
            'Sensor': sensor,
            'Chemical_1': chem1,
            'Chemical_2': chem2,
            'KS_statistic': stat,
            'p_value': p
        })

ks_df = pd.DataFrame(ks_results)
os.makedirs('results', exist_ok=True)
ks_df.to_csv('results/ks_test_results.csv', index=False)

print("Kolmogorov-Smirnov test results saved to: results/ks_test_results.csv")
ks_df.head()
```

```python
kruskal_df = pd.read_csv('results/kruskal_result.csv', index_col=0)
mw_df = pd.read_csv('results/mannwhitney_results.csv')
ks_df = pd.read_csv('results/ks_test_results.csv')
spearman_corr = pd.read_csv('results/spearman_correlation.csv', index_col=0)

fig, axs = plt.subplots(3, 2, figsize=(18, 16))
fig.suptitle('Summary of Statistical Analysis', fontsize=20)

inter_group_var = 483.1982
intra_group_var = 6.1823
axs[0, 0].bar(['Inter-group', 'Intra-group'], [inter_group_var, intra_group_var], color=['steelblue', 'orange'])
axs[0, 0].set_title('Inter vs Intra-group Variance')
axs[0, 0].set_ylabel('Average Variance')

axs[0, 1].hist(kruskal_df['p_value'], bins=30, color='purple', edgecolor='black')
axs[0, 1].set_title('Kruskal-Wallis p-value Distribution')
axs[0, 1].set_xlabel('p-value')
axs[0, 1].set_ylabel('Number of Sensors')

mw_top = mw_df.sort_values('p_value').head(10)
axs[1, 0].barh(
    [f"{r['Chemical_1']} vs {r['Chemical_2']}" for _, r in mw_top.iterrows()],
    mw_top['p_value'],
    color='green'
)
axs[1, 0].invert_yaxis()
axs[1, 0].set_title('Top 10 Mann-Whitney Pairs')
axs[1, 0].set_xlabel('p-value')

ks_top = ks_df.sort_values('p_value').head(10)
axs[1, 1].barh(
    [f"{r['Chemical_1']} vs {r['Chemical_2']}" for _, r in ks_top.iterrows()],
    ks_top['p_value'],
    color='crimson'
)
axs[1, 1].invert_yaxis()
axs[1, 1].set_title('Top 10 Kolmogorov-Smirnov Pairs')
axs[1, 1].set_xlabel('p-value')

sns.heatmap(spearman_corr, ax=axs[2, 0], cmap='coolwarm', square=True, cbar_kws={'label': 'Spearman ρ'})
axs[2, 0].set_title('Spearman Correlation Between Chemical Profiles')

axs[2, 1].axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
from sklearn.manifold import TSNE

sensor_data_only = df.drop(columns=['sample', 'Chemical', 'Replicate'])

sensor_data_only = sensor_data_only.dropna()

chemical_labels = df.loc[sensor_data_only.index, 'Chemical']

scaler = StandardScaler()
sensor_scaled = scaler.fit_transform(sensor_data_only)

tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
tsne_result = tsne.fit_transform(sensor_scaled)

tsne_df = pd.DataFrame({
    'TSNE1': tsne_result[:, 0],
    'TSNE2': tsne_result[:, 1],
    'Chemical': chemical_labels.values
})

plt.figure(figsize=(12, 8))
sns.scatterplot(data=tsne_df, x='TSNE1', y='TSNE2', hue='Chemical', palette='tab20', s=80)
plt.title('t-SNE Visualization of Chemical Sensor Responses (Nonparametric)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
import matplotlib.patches as patches
from matplotlib.patches import Ellipse

def confidence_ellipse(x, y, ax, n_std=2.0, facecolor='none', edgecolor='black', **kwargs):
    """
    Create a plot of a covariance confidence ellipse of x and y
    """
    if x.size != y.size:
        raise ValueError("x and y must be the same size")
    
    cov = np.cov(x, y)
    pearson = cov[0, 1]/np.sqrt(cov[0, 0] * cov[1, 1])
    ell_radius_x = np.sqrt(1 + pearson)
    ell_radius_y = np.sqrt(1 - pearson)
    ellipse = Ellipse((np.mean(x), np.mean(y)),
                      width=ell_radius_x * n_std * 2,
                      height=ell_radius_y * n_std * 2,
                      facecolor=facecolor,
                      edgecolor=edgecolor,
                      **kwargs)
    ax.add_patch(ellipse)

tsne_df = pd.DataFrame({
    'TSNE1': tsne_result[:, 0],
    'TSNE2': tsne_result[:, 1],
    'Chemical': chemical_labels.values
})

plt.figure(figsize=(14, 10))
ax = sns.scatterplot(data=tsne_df, x='TSNE1', y='TSNE2', hue='Chemical', palette='tab20', s=80)

for chem in tsne_df['Chemical'].unique():
    subset = tsne_df[tsne_df['Chemical'] == chem]
    confidence_ellipse(subset['TSNE1'], subset['TSNE2'], ax, n_std=2.5, edgecolor='black', alpha=0.3)

plt.title('t-SNE Visualization of Chemical Sensor Responses (with Group Ellipses)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

# Final Conclusions

This project investigated the ability of a colorimetric sensor array to distinguish between different toxic industrial chemicals (TICs) based on rapid (2-minute) exposure readings, using a strictly nonparametric analytical framework.

### Key Findings:
- **Variance Analysis**:  
  The average inter-chemical variance was nearly **80 times greater** than the average intra-chemical variance. This suggests that chemical identities induce reliably distinct sensor response profiles, far exceeding the noise or variability within replicates.

- **Kruskal-Wallis H-Test**:  
  The majority of sensors showed statistically significant differences across chemical groups (**p-values < 0.05**), supporting strong chemical discrimination capabilities at the sensor level.

- **Pairwise Testing (Mann-Whitney and Kolmogorov-Smirnov Tests)**:  
  Specific chemical pairs were identified as significantly distinguishable based on both location (Mann-Whitney) and distribution shape (KS-test) metrics.

- **Spearman Correlation Analysis**:  
  The pattern similarity analysis revealed strong clustering among chemically similar compounds and clear separation from chemically distinct groups.

- **t-SNE Nonparametric Visualization**:  
  The t-SNE mapping showed that replicates from the same chemical group cluster tightly, with minimal overlap between different groups. This provides strong, nonparametric visual evidence of effective chemical discrimination.

### Overall Conclusion:
The results consistently demonstrate that the colorimetric sensor array can effectively and reliably distinguish between different toxic industrial chemicals after only 2 minutes of exposure.  
This supports the potential use of such arrays for rapid, portable chemical detection applications.
